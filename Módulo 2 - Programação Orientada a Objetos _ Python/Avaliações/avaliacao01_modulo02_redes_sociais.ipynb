# Avaliação módulo 1 - Degree Data Science

Na empresa em que você trabalha foi solicitada a implementação de um sistema de cadastro dos fornecedores. As informações que devem ser cadastradas são:

- Código do fornecedor;
- Nome do fornecedor;
- Telefone do fornecedor;
- Email do fornecedor.

Para isso, você deverá criar um script para armazenar essas informações em alguma estrutura de dados.

Também como parte do sistema de cadastro, é necessário criar uma "interface de navegação" pelo sistema, que tenha as seguintes funcionalidades:

- Permite a exibição (usando o `print`) de todas as informaçoes de contatos específicos, a partir da busca pelo código do fornecedor;
- Permite a remoção e adição de fornecedores à estrutura de cadastro.

Requisitos:

- Requisito #1 - Não existe um limite para fornecedores cadastrados, portanto você deverá permitir que o usuário cadastre a quantidade que desejar.
- Requisito #2 - Crie pelo menos uma função em seu script. Você deverá escolher pelo menos uma tarefa do seu script e implementar em uma função para chamá-la dentro da sua lógica de cadastro.
- Requisito #3 - Crie pelo menos uma verificação em seu código, por exemplo: Se o dicionário estiver vazio, não permita a opção remover.

Tópicos avaliados:

- Variáveis;
- Tipos de dados;
- Operadores aritméticos;
- I/O de dados;
- Expressões lógicas e condicionais;
- Laços de repetição;
- Listas e dicionários;
- Funções.


```python
#função para listar os fornecedores cadastrados

#Aqui, utilizo a desconstrução de tupla, que possui as quatro informações do fornecedor, para 4 variáveis distintas; 
#assim, o código ficará muito mais legível, pois é mais fácil entender. 
#E para exibir as informações, 
#utilizei as f-strings, do Python 3.6 - onde, por exemplo, {nome} será 
#substituído pelo valor de nome.

def listar(fornecedor):
    print(f"\nA seguir a lista de fornecedores já cadastrados:")
    for mostrar in fornecedor:
        cod_forn, nome_forn, tel_forn, email_forn = mostrar
        print(f"Código: {cod_forn}, Nome: {nome_forn}, Telefone: {tel_forn}, e-mail: { email_forn}")
```

```python
#função para cadastrar fornecedor

#criando uma tupla com os dados, note a presença de dois parenteses no append

def cadastar(fornecedor):
    cod_forn = int(input("\nInsira o novo código do fornecedor: "))
    nome_forn = input ("Insira o nome do fornecedor: ")
    tel_forn = input ("Insira o telefone do fornecedor: ")
    email_forn = input ("Insira o e-mail do fornecedor: ")
    fornecedor.append((cod_forn,nome_forn, tel_forn, email_forn))
    print(f"\nFornecedor com código {cod_forn} adicionado com sucesso")
```

```python
#função para buscar fornecedor

def buscar(fornecedor):
    cod_forn_procurado = int(input("Código fornecedor a ser localizado: "))
    for mostrar in fornecedor:
        cod_forn, nome_forn, tel_forn, email_forn = mostrar
        if cod_forn == cod_forn_procurado:
            print(f"\nCódigo: {cod_forn}, Nome: {nome_forn}, Telefone: {tel_forn}, e-mail: { email_forn}")
            break
#        else:
#            print(f"\nFornecedor com código {cod_forn_procurado} não localizado")
```

```python
def deletar(fornecedor):
    cod_forn_deletar = int(input("Código fornecedor a ser removido: "))
    for deletar in fornecedor:
        cod_forn, nome_forn, tel_forn, email_forn = deletar
        if cod_forn == cod_forn_deletar:
            cod_forn_deletar = cod_forn_deletar - 1
            del fornecedor[cod_forn_deletar]
            print(f"\nFornecedor com código {cod_forn_deletar+1} deletado")
            break
#        else:
#            print(f"\nFornecedor com código {cod_forn_deletar} não localizado")
```

```python
fornecedor = []
    
while True:
    print("\n1 - Cadastrar fornecedor ")
    print("2 - Listar cadastro: ")
    print("3 - Procurar fornecedor: ")
    print("4 - Remover fornecedor: ")
    opcao_usuario = int(input("\nPor favor, escolha uma das opções de 1 a 4: "))
    
    if opcao_usuario == 1:
        cadastar(fornecedor)
    
    elif opcao_usuario == 2:
        listar(fornecedor)
    
    elif opcao_usuario == 3:
        buscar(fornecedor)
    
    elif opcao_usuario == 4:
        if len(fornecedor) == 0:
            print("\nNão há fornecedor(es) cadastrado(s)")
        else:
            deletar(fornecedor)
    else:
        print("\nOpção inválida")
```
